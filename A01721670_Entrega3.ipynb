### Problematica a resolver:
-La evaluación PISA es una evaluación utilizada en ámbito nacional. Dicha evaluación sirve, de cierta manera, para medir la calidad de la educación en México. Dicho programa proporcionará más oportunidades para repasar, estudiar y preparar. Fases claves para que el alumno vaya aprendiendo y mejorando. El programa será altamente interactivo y personalizado. Tendrá un menú con seis opciones: alta de preguntas de prueba PISA, actualizar preguntas de la prueba PISA, estudiar preguntas de cierta área, presentar un quiz, reporte de calificaciones y salir.
```python
def estudiar_area(nombre):

  lista_mate = []
  lista_ciencias = []
  lista_lect = []
  num_preguntas = int(input("Ingresa cuantas preguntas tendra el quiz: "))
  area = input("Ingresa el area que deseas estudiar: ")

  if area == "Matematicas":
    for pregunta in preguntas:
      if pregunta[1] == "mate":
        lista_mate.append(pregunta)
    for num in range(num_preguntas):
        print(lista_mate[num])
  
  elif area == "Ciencias":
    for pregunta in preguntas:
      if pregunta[1] == "ciencias":
        lista_ciencias.append(pregunta)
    for num in range(num_preguntas):
        print(lista_ciencias[num])

  elif area == "Lectura":
    for pregunta in preguntas:
      if pregunta[1] == "lectura":
        lista_lect.append(pregunta)
    for num in range(num_preguntas):
        print(lista_lect[num])

  else:
    print("Area invalida")
      

estudiar_area(nombre_archivo)
```

```python
def presentar_quiz(nombre):

  global pregunta
  global total_correctas
  global total_incorrectas

  total_correctas = 0
  total_incorrectas = 0
  total_preguntas = 0
  import time

  start = time.time()

  print("Presentar quiz, por favor responde: \n")

  for pregunta in preguntas: 
    while total_preguntas != 10:
      print(f"{pregunta[2]}\n {pregunta[3]:>25} {pregunta[4]:>25} {pregunta[5]:>25} {pregunta[6]:>25}\n")
      respuesta = input("\nRespuesta: \n")

      if respuesta == pregunta[7]:
        total_correctas = total_correctas + 1

      else:
        total_incorrectas = total_incorrectas + 1

      total_preguntas = total_preguntas + 1
      break
  
  end = time.time()
  tiempo = end - start
  print(f"\nTiempo: {tiempo}")

  if tiempo <= 1200:
    print("Tiempo aprobado")
  else:
    print("Tiempo rechazado")

      
presentar_quiz(nombre_archivo)
```
